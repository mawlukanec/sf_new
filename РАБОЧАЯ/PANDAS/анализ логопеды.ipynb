```python
import pandas as pd
logopedi_data = pd.read_csv('d:/python/GUESS/РАБОЧАЯ/PANDAS/data/logopedi_1.03_12.04.csv', sep=';', encoding='windows 1251')
logopedi_df = logopedi_data.copy()

# удаление последней строки с обобщенной информацией и NaN
dell = logopedi_df.shape[0]
logopedi_df = logopedi_df[0:dell-1] 
# print(logopedi_df)

# приведение времени
logopedi_df['Дата'] = pd.to_datetime(logopedi_df['Дата'])# , dayfirst= True)

def split_gr(gr):
    temp = gr.split(' ')[-1] # нужно обрезать список!!!
    return temp

logopedi_df['группа']= logopedi_df['Запись'].apply(split_gr) 
logopedi_df['группа'] = logopedi_df['группа'].astype('int8')
#print(logopedi_df.tail(3))
#display(logopedi_df.dtypes)

# срез данных
# logoped = logopedi_df.iloc[:, 7:].groupby(by= 'группа').sum().copy()
# logoped['k'] = (logoped['Потрачено, руб.']/logoped['Выполненные заявки'])/logoped['Переходы по ссылкам']
#display(logoped.sort_values(by= 'k', ascending= True).head(20))


logopedi_df['k'] = (logopedi_df['Потрачено, руб.']/logopedi_df['Выполненные заявки'])/logopedi_df['Переходы по ссылкам']
logopedi_df.groupby(['группа','Запись', 'k'])['Потрачено, руб.'].sum()

def split_gr(gr):
    temp = gr.split(' ')[0] # нужно обрезать список!!!
    temp = gr.split('_')[:2]
    temp = '_'.join(temp)    
    return temp
logopedi_df['Запись_без №']= logopedi_df['Запись'].apply(split_gr) 

#                   МЕТОД GROUPBY КАК СПОСОБ ПОСТРОЕНИЯ СВОДНЫХ ТАБЛИЦ
logopeds = logopedi_df.groupby(['группа','Запись_без №'])['Потрачено, руб.'].sum().unstack().copy()
logopeds.fillna(value = 0, inplace = True)
logopeds

# или                МЕТОД PIVOT_TABLE ДЛЯ ПОСТРОЕНИЯ СВОДНЫХ ТАБЛИЦ

logo = logopedi_df.pivot_table(
    values='Потрачено, руб.',
    index='группа',
    columns='Запись_без №',
    fill_value=0
).round()
# display(logoped.sort_values(by= 'Комментарии', ascending= False).head(20))
logo['summ'] = logo.sum('columns')
logo

```
