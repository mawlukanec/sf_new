```python
import pandas as pd
metrika = pd.read_csv('C:/ANNA/! Посевы/анализы/логопеды/onlineschoollogo.com-Метки_UTM.csv', sep = ',')
vk = pd.read_csv('C:/ANNA/! Посевы/анализы/логопеды/posts_stats.csv',  sep=';', encoding='windows 1251')
metrika_a = metrika.copy()
logopedi_df = vk.copy()

#                       метрика 

mask = metrika_a['UTM Medium'] == 'Orange'
mask_vk = metrika_a['UTM Source'] == 'vk_market'
x = metrika_a[mask&mask_vk].copy()

def groups(gp):
    temp = gp.split('_')    
    return (temp[-1])
x['grooups'] = x['UTM Campaign'].apply(groups)
# x1 = x.groupby(
#     by=['grooups','UTM Content'], 
#     as_index=False
# )['Целевые визиты ([PRO] Рега на марафон запуск 2 - JS Отправка формы)'].sum().unstack()

# for i in x1.columns:
#     x1[i] = x1[i].fillna(0)

x1 = x.pivot_table(
    values = 'Целевые визиты ([PRO] Рега на марафон запуск 2 - JS Отправка формы)',
    index = ['grooups'],
    columns = 'UTM Content',
    aggfunc='median',    
    fill_value=0
)
x1['реги_сумм'] = x1.sum('columns')
# x1.sort_values('реги_сумм', ascending=False, inplace=True)
x1.reset_index(inplace= True)  # переносим индексы групп в признак
# x1.to_excel('C:/ANNA/! Посевы/анализы/логопеды/onlineschoollogo.xlsx')

#                          VK

# удаление последней строки с обобщенной информацией и NaN
dell = logopedi_df.shape[0]
logopedi_df = logopedi_df[0:dell-1] 
# приведение времени
logopedi_df['Дата'] = pd.to_datetime(logopedi_df['Дата'])

def split_gr(gr):
    temp = gr.split(' ')[-1] # нужно обрезать список!!!
    return temp

logopedi_df['grooups']= logopedi_df['Запись'].apply(split_gr) 
# logopedi_df['grooups'] = logopedi_df['grooups'].astype('int8')
logopedi_df['k'] = (logopedi_df['Потрачено, руб.']/logopedi_df['Выполненные заявки'])/logopedi_df['Переходы по ссылкам']
logopedi_df.groupby(['grooups','Запись', 'k'])['Потрачено, руб.'].sum()

def split_gr(gr):
    temp = gr.split(' ')[0] # нужно обрезать список!!!
    temp = gr.split('_')[:2]
    temp = '_'.join(temp)    
    return temp
logopedi_df['Запись_без №'] = logopedi_df['Запись'].apply(split_gr) 

# print(logopedi_df.head())

# #                   МЕТОД GROUPBY КАК СПОСОБ ПОСТРОЕНИЯ СВОДНЫХ ТАБЛИЦ
# logopeds = logopedi_df.groupby(['grooups','Запись_без №'])['Потрачено, руб.'].sum().unstack().copy()
# logopeds.fillna(value = 0, inplace = True)
# logopeds

# или                МЕТОД PIVOT_TABLE ДЛЯ ПОСТРОЕНИЯ СВОДНЫХ ТАБЛИЦ

# print(logopedi_df.head())

logo = logopedi_df.pivot_table(
    values=['Потрачено, руб.','Выполненные заявки','Переходы по ссылкам'],
    index='grooups',
    columns='Запись_без №',
    fill_value=0
).round(0)


# display(logoped.sort_values(by= 'Комментарии', ascending= False).head(20))
logo['затраты_сумм'] = logo['Потрачено, руб.'].sum('columns')
logo.reset_index(inplace= True)
print(logo.head())

logo_x = logo.copy()

# display(x1.head(5),'\n', logo.head(5))
# merged = logo_x.merge(
#     x1,    
#     on = 'grooups',
#     how = 'left',
#     suffixes= ('_vk', '_met')
#     )
# merged['цена реги'] = merged['затраты_сумм']/merged['реги_сумм']
# merged.sort_values('цена реги', ascending= True, ignore_index=True)
```
