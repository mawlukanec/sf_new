```python
# Задание №1
input_string1 = input('Введите 1-ую последовательность идентификаторов: ')
input_string2 = input('Введите 2-ую последовательность идентификаторов: ')
# ваш код здесь

# удаляем все пробелы и затем разделяем введенные числа (разделитель - ',') для каждой последовательности 
input_string1 = input_string1.replace(' ', '').split(',')
input_string2 = input_string2.replace(' ', '').split(',')
# поиск ошибки
try:
  # создаём пустые списки для каждой последовательности
  input_string1_int = []; input_string2_int = []
  # цикл по элементам введенных последовательностей чисел с преобразованием из в целое число соответственно
  for i in range(len(input_string1)):
    i_1 = int(input_string1[i])
    input_string1_int.append(i_1)    
  for y in range(len(input_string2)):
    i_2 = int(input_string2[y])
    input_string2_int.append(i_2)   
  # если на каком то этапе появляется ошибка ValueError печатаем ошибку  'Некорректный ввод' и завершаем выполнение программы
except ValueError:
  print('Некорректный ввод')
# если ошибок нет, то продолжаем выполнение кода 
else:
  # функция для поиска пересечений в списках
  def pure_intersection(input_string1, input_string2):
    # создаём пустой список  
    result = []
    # преобразуем полученные списки в множества соответственно:
    in_1 = set(input_string1_int); in_2 = set(input_string2_int)
    # ищем пересечения множеств
    result = list(in_1.intersection(in_2)) 
    # возвращаем результат работы функции - список пересечений, если они есть. если нет - возвращаем пустой список   .
    return result
  
  print(pure_intersection(input_string1, input_string2))
```

```python
# Задание №2
input_string = input('Введите последовательность чисел: ')

# фунцкия для нахождения минимума и максимума в введнной последовательности чисел
def find_min_max(input_string):
    # объявляем глобальные переменные для минимального и максимального значения соответственно
    global min_value; global max_value
    # заменяем все ',' на '.', затем разделяем строку по пробеллам. в результате получаем список модифицированных элементов из строки
    input_string_list = (input_string.replace(',', '.').split(' '))
    # инициализируем пустой список для проверки корректности ввода (игнорируем буквенный ввод числа)
    result = []
    # цикл для проверки корректности ввода 
    for i in input_string_list:
    # проверка i-того элемента списка input_string_list на то, состоит ли он только из букв, и если да
        if i.isalpha() != True:
        # если нет, то добавляем к списку result i-тый элемент списка input_string_list
            result.append(i)
    # инициализируем пустой список для преобразования типа данных str в int или float
    result_list = []
    # цикл для преобразования типа данных str в int или float
    for i in result:
        # проверяем на возможную ошибку в преобразовании типов данных
        try:
            # если значение сисла будет одинаково независимо от типа данных int or float
            if float(i) == int(i):
                # добавляем к результирующему списку int число
                result_list.append(int(i))
        # если возникла ошибка
        except ValueError:
            # добавляем к результирующему списку float число
            result_list.append(float(i))
    # сообщаем минимальное и максимальное значение(в соответствии с типом данных) переменным min_value, max_value соответственно
    min_value = min(result_list); max_value =  max(result_list)
    # возвращаем минимальное и максимальное значение из введенной строки 
    return min_value, max_value
# сообщаем tuple_output результат выполнения функции find_min_max()
tuple_output = find_min_max(input_string)
# выводим на экран минимальное и максимальное значение из введенной строки 
print('Minimum: ', tuple_output[0])
print('Maximum: ', tuple_output[1])
```

```python
# Задание №3

input_string = input('Введите последовательность чисел: ')
# функция пдля рассчета медианы введенной последовательеости чисел
def find_median(input_string):
    # создаем пкстой список
    result = []
    # если пустой список = введенному, т.е. пустой - выводим на экран фразу 'Некорректный ввод.'
    if result == list(input_string):
        return 'Некорректный ввод.'
        
    # разделяем введенную строку по запятой и пробелу 
    input_string_list = input_string.split(', ')
    # отлавливаем ошибку - неверно введенное число
    try:
        # цикл по элементам списка
        for i in input_string_list:           
            # преобразуем элемент строки в число (float)
            i = float(i)
            # добавляем в result получившиеся числа
            result.append(float(i))
    # если элемент списка нельзя преобразовать в число
    except ValueError:
        # выводим на экран фразу 'Некорректный ввод.'
        return 'Некорректный ввод.'
    # если преобразование прошло без оштбок 
    else:
        # сортируем список по возрастанию
        result.sort()
        # функция нахождения медианы
        def median(result):
        # цикл по элементам списка result
            for i in result:
            # если длинна списка > 2
                if len(result) > 2:
                # удаляем из списка крайние элементы
                   result = result[1:-1]
      # если длинна списка = 1, это и будет медианой введенной строки 
            if len(result) == 1:
            # выводим на экран [0] элемент строки 
                return result[0]
            else:
                # если длинна списка = 2, то складываем оставшиеся элементы списка и делим на 2
                return (result[0] + result[1])/2
        # сообщаем перменной result результат выполнения функции median(result)
        result = median(result)
        # возвращаем медиану строки в формате  'Median: ' + строка со значением вычисленной медианы введенной строки   
        return 'Median: '+ str(result)
print(find_median(input_string))
```

```python
# Задание №5  
n = input('Введите число: ')

def decompose_factorial(n):
    # преобразуем введенную строку в число  с типом данных int
    n = int(n)
    # функция нахождения факториала числа
    def factorial(num):
        # остановка рекурсии при num = 1 или 0 и возращаем 1
        if num == 0 or num ==1:
            return 1
        # дальнейшее продолжение рекурсии
        return num*factorial(num - 1)
    # сообщаем переменной  n_factorial результат выполнения рекурсивной функции factorial(n) 
    n_factorial = factorial(n)
    # функция нахождения всех простых множителей факториала введенного числа
    def prime_factorization(value):
        # пустой список
        temp_list = []
        # сообщаем переменной d начальное значение 2 (имнимальное значение простого числа (после 1))
        d = 2 
        # пока d^2 будет меньше или = факториалу введенного числа
        while d * d <= value:
            # если остаток от деления на d равен 0
            if value % d == 0:
                # добавляем значение d к списку множителей
                temp_list.append(d)
                # делим значение переменной value на d и присваиваем переменной value (остаток от деления) результат
                value //= d
            # если остаток от деления value на d неравен 0 
            else:
                # увеличиваем значение d на 1
                d += 1
        # если значение d больше 1
        if value > 1:
            # добавляем значение к списку  temp_list
            temp_list.append(value)
        # возвращаем список простых множителей       
        return temp_list
    # сообщаем переменной temp_list список простых множителей
    temp_list = prime_factorization(n_factorial)
    # фунция сортировки простых множителей по возрастанию
    def sort_dict(temp_list):
        # сосдаем пустой словать
        temp_dict = {}
        # цикл по уникальным значениям из списка простых множителей
        for i in set(temp_list):
            # записываем в словарь по ключу i количество вхождений ключа в список простых множителей
            temp_dict[i] = temp_list.count(i)
            # сортируем словарь при помощи функции sorted() по значению ключа и преобразуем получившийся кортеж обратно в словарь
            sort_dict = dict(sorted(temp_dict.items(), key = lambda x: x[0]))
        # возвращаем отсортированный словарь, где ключ - уникальное значение простого множителя, значение - количество входжений 
        # данного множителя в список простых множителей  
        return sort_dict
    # сообщаем sort_dict результат выполнения функции sort_dict(temp_list) 
    sort_dict = sort_dict(temp_list)
    # инициализируем пустой список
    temp_list = []
    # цикл по ключам отсортированного словаря
    for i in sort_dict:
        # 
        # temp_str = ''
        # если количество вхождений простого множителя = 1
        if sort_dict[i] == 1:
            # добаляем к списку только сам множитель с типом данных str
            temp_list.append(str(i))
        # если оличество вхождений простого множителя > 1
        else:
            # создаем строку из самого множителя, знака степени и количества вхождений множителя в список простых множителей    
            temp_str = str(i) + '^' + str(sort_dict[i])
            # добавляем к итоговому списку строку temp_str
            temp_list.append(temp_str)
    # возвращаем строку из произведений простых множителей возведенных в соответствующую степень
    return ' * '.join(temp_list)
print(decompose_factorial(n))
```

```python
# Задание №6 
s = input('Введите строку: ')

import string
alphabet = string.ascii_lowercase

def replace_duplicates(temp):
    # запускаем бесконечный цыкл
    while True:
        # сравниваем длинну введенной строки с множеством этой же строки для определения наличия повторяющихся элементов
        if len(temp) == len(set(temp)):
            # если повторяющихся элементов нет - завершаем бесконечный цикл
            break
        # если есть повторения элементов 
        else:
            # цикл по элементам строки множества
            for i in str(set(temp)):
                # инициализируем пустую строку
                index = []
                # инициализируем переменные nums_index и  duble для подсчета индека повторяющегося элемента и определения его двойного повторения соответственно
                nums_index, duble = 0, 0             
                # если буква есть в списке temp 
                for letter in temp:
                    # увеличиваем переменную nums_index для подсчета индека повторяющегося элемента на 1
                    nums_index += 1
                   # если элемент строки множества равен элементу в строке temp 
                    if i == letter:
                        # увеличиваем duble на 1
                        duble +=1
                        # добавляем индекс поторяющегося элемента
                        index.append(nums_index)
                        # если число совпадений  = 2
                        if duble == 2:
                            # сообщаем litera и letter_indexes соответственно дублирующуюся букву и список её индексов
                            litera = i
                            letter_indexes = index    
                            # завершаем работу цикла                      
                            break
            # цикл по индексам alphabet
            for i in range(len(alphabet)):
                # если буква из alphabet соответствует дублирующейся букве
                if alphabet[i] == litera:
                    # записываем в переменную chenge_litera следующую букву из alphabet, при этом если дублирующейся буквой являестя 'z' то берем первую букву из alphabet  
                    if alphabet[i] == 'z':
                        chenge_litera = alphabet[0]  
                    else:
                        chenge_litera = alphabet[i+1]  
                    # завершаем цикл              
                    break
            # добавляем в конец строки заменяющую дубль букву (chenge_litera) согласно alphabet    
            temp += chenge_litera
            # преобразуем строку в список для исрользования метода .pop()            
            temp = list(temp)
            # удаляем дублирующийся элемент в обратном порядке
            temp.pop(letter_indexes[1]-1); temp.pop(letter_indexes[0]-1)
    # возвращаем получившийся список в виде строки      
    return  ''.join(temp)

print(replace_duplicates(s))
```
